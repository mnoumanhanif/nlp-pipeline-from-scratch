## Task P1T1: HTML to Plain Text Utility


```python
import urllib.request
from bs4 import BeautifulSoup

def fetch_clean_text(url: str) -> str:
    """
    Downloads HTML from a URL and extracts plain text.
    """
    # 1) Download HTML using urllib.request
    # Adding a User-Agent helps avoid 403 Forbidden errors on some web pages
    req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    with urllib.request.urlopen(req) as response:
        html_bytes = response.read()
        html = html_bytes.decode('utf-8', errors='ignore')

    # 2) Extract text using BeautifulSoup(html).get_text()
    soup = BeautifulSoup(html, 'html.parser')
    # Using separator=' ' prevents words from merging when HTML tags are stripped
    plain_text = soup.get_text(separator=' ', strip=True)

    # 3) Return plain text
    return plain_text

# ==========================================
# Deliverable Execution
# ==========================================
if __name__ == "__main__":
    # Using the MIT CSAIL people page as suggested in the assignment
    target_url = "https://www.csail.mit.edu/people/"

    # Fetching the raw HTML just to satisfy the "before cleaning" print requirement
    req = urllib.request.Request(target_url, headers={'User-Agent': 'Mozilla/5.0'})
    with urllib.request.urlopen(req) as response:
        raw_html = response.read().decode('utf-8', errors='ignore')

    # Calling our utility function
    cleaned_text = fetch_clean_text(target_url)

    print("--- BEFORE CLEANING (First 500 characters) ---")
    print(raw_html[:500])
    print("\n" + "="*50 + "\n")
    print("--- AFTER CLEANING (First 500 characters) ---")
    print(cleaned_text[:500])
```

## Task P1T3: Text Normalization Pipeline
We will apply lowercasing, replace all numbers with a `<NUM>` tag, remove punctuation, and strip out any extra whitespaces.
* **Replacing digits with `<NUM>`:** We do this to reduce vocabulary sparsity. In language modeling, the exact value of a number (e.g., "2026" vs "2025") often matters less than the fact that a number appeared. Grouping them under `<NUM>` helps the model learn better statistical probabilities.
* **Removing Punctuation:** We remove punctuation so that words attached to punctuation (like `"research,"` and `"research"`) are treated as the exact same token by our Unigram and Bigram models later. This significantly reduces Out-Of-Vocabulary (OOV) errors and shrinks the vocabulary size.

```python
import re
import string

def normalize_text(text: str) -> str:
    # 1. Lowercasing
    norm_text = text.lower()

    # 2. Replace digits with <NUM> (reduces sparsity)
    norm_text = re.sub(r'\d+', '<NUM>', norm_text)

    # 3. Remove punctuation (standardizes tokens)
    # This translates any punctuation character into 'None' (removes it)
    norm_text = norm_text.translate(str.maketrans('', '', string.punctuation))

    # 4. Remove extra whitespace (cleans up gaps left by removed punctuation)
    norm_text = re.sub(r'\s+', ' ', norm_text).strip()

    return norm_text

# Apply normalization to the text extracted from P1T1/P1T2
normalized_text = normalize_text(extracted_text)

# ==========================================
# Deliverables Output
# ==========================================
print("=== TEXT NORMALIZATION PIPELINE ===")
print("Rules Applied:")
print("1. Lowercasing: To treat 'MIT' and 'mit' as the same word.")
print("2. Replaced digits with <NUM>: To reduce vocabulary sparsity.")
print("3. Removed punctuation: To prevent tokens like 'word,' and 'word' from splitting.")
print("4. Removed extra whitespace: To ensure clean token boundaries.\n")

print("--- BEFORE NORMALIZATION (First 300 chars) ---")
print(extracted_text[:300])
print("\n" + "="*50 + "\n")
print("--- AFTER NORMALIZATION (First 300 chars) ---")
print(normalized_text[:300])
```

## Task P1T4: Stopwords and Corpus Statistics

* **Total Tokens:** Notice how drastically the total token count drops after removing stopwords.
* **Type-Token Ratio (TTR):** The TTR should increase noticeably after stopword removal. Because stopwords are highly repetitive, removing them reduces the total token count much faster than it reduces the unique vocabulary size, resulting in a "richer" text representation.

```python
import nltk
from collections import Counter
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download('stopwords')

# 1. Tokenize the normalized text from P1T3
# Since we already removed punctuation and extra whitespaces, .split() is perfect here
tokens_before = normalized_text.split()

# ---------------------------------------------------------
# 2. Compute Statistics BEFORE Stopword Removal
# ---------------------------------------------------------
total_tokens_before = len(tokens_before)
vocab_before = set(tokens_before)
vocab_size_before = len(vocab_before)
ttr_before = vocab_size_before / total_tokens_before if total_tokens_before > 0 else 0

freq_before = Counter(tokens_before)
top_25_before = freq_before.most_common(25)

# ---------------------------------------------------------
# 3. Remove English Stopwords
# ---------------------------------------------------------
stop_words = set(stopwords.words('english'))
tokens_after = [word for word in tokens_before if word not in stop_words]

# ---------------------------------------------------------
# 4. Compute Statistics AFTER Stopword Removal
# ---------------------------------------------------------
total_tokens_after = len(tokens_after)
vocab_after = set(tokens_after)
vocab_size_after = len(vocab_after)
ttr_after = vocab_size_after / total_tokens_after if total_tokens_after > 0 else 0

freq_after = Counter(tokens_after)
top_25_after = freq_after.most_common(25)

# ==========================================
# Deliverables Output
# ==========================================
print("=== STATISTICS SUMMARY TABLE ===")
print(f"{'Metric':<30} | {'Before Stopwords':<20} | {'After Stopwords':<20}")
print("-" * 78)
print(f"{'Total Token Count':<30} | {total_tokens_before:<20} | {total_tokens_after:<20}")
print(f"{'Vocabulary Size (Unique)':<30} | {vocab_size_before:<20} | {vocab_size_after:<20}")
print(f"{'Type-Token Ratio (TTR)':<30} | {ttr_before:<20.4f} | {ttr_after:<20.4f}")
print("\n")

print("=== TOP 25 MOST FREQUENT TOKENS ===")
print(f"{'Rank':<5} | {'Before Stopwords':<25} | {'After Stopwords':<25}")
print("-" * 65)

for i in range(25):
    # Safely get the word and count, formatting them as "word (count)"
    word_before = f"{top_25_before[i][0]} ({top_25_before[i][1]})" if i < len(top_25_before) else "-"
    word_after = f"{top_25_after[i][0]} ({top_25_after[i][1]})" if i < len(top_25_after) else "-"

    print(f"{i+1:<5} | {word_before:<25} | {word_after:<25}")
```

## Task P1T5: POS Tagging + POS Distribution
It applies `nltk.pos_tag()` directly to the normalized tokens from P1T3. It then uses Python's `Counter` to calculate the distribution of the tags and extracts unique examples of nouns (tags starting with `NN`) and verbs (tags starting with `VB`).

Since we ran the POS tagger on the normalized text (which is completely lowercased), we might notice in our results that NLTK struggles slightly with Proper Nouns (NNP). Standard POS taggers heavily rely on capital letters to identify names and locations. Because we lowercased "MIT" to "mit", the tagger might classify it as a standard noun (NN) instead of a proper noun (NNP).

```python
import nltk
from collections import Counter

# Ensure the POS tagger models are downloaded
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng') # For newer NLTK versions

# 1. Tokenize the normalized text from P1T3
# Using .split() since the text is already lowercased and stripped of punctuation
normalized_tokens = normalized_text.split()

# 2. Perform POS tagging on normalized tokens
pos_tags = nltk.pos_tag(normalized_tokens)

# 3. Calculate POS distribution (frequencies)
# We extract just the tag from the (word, tag) tuple to count them
tag_counts = Counter(tag for word, tag in pos_tags)
top_10_tags = tag_counts.most_common(10)

# 4. Extract Example Nouns and Verbs
# Noun tags in Penn Treebank start with 'NN' (NN, NNS, NNP, NNPS)
# Verb tags in Penn Treebank start with 'VB' (VB, VBD, VBG, VBN, VBP, VBZ)
unique_nouns = set()
unique_verbs = set()

for word, tag in pos_tags:
    if tag.startswith('NN'):
        unique_nouns.add(word)
    elif tag.startswith('VB'):
        unique_verbs.add(word)

# Grab exactly 15 examples of each
example_nouns = list(unique_nouns)[:15]
example_verbs = list(unique_verbs)[:15]

# ==========================================
# Deliverables Output
# ==========================================
print("=== POS DISTRIBUTION TABLE (Top 10) ===")
print(f"{'Rank':<6} | {'POS Tag':<10} | {'Frequency':<10}")
print("-" * 35)
for i, (tag, count) in enumerate(top_10_tags):
    print(f"{i+1:<6} | {tag:<10} | {count:<10}")

print("\n" + "="*50 + "\n")

print("=== EXAMPLE NOUNS (15) ===")
# Formatting as a comma-separated string for a cleaner report
print(", ".join(example_nouns))

print("\n=== EXAMPLE VERBS (15) ===")
print(", ".join(example_verbs))
```

## Task P1T6: Lemmatization vs Stemming
It selects 125 unique tokens from your normalized text, applies all three techniques, and prints the perfectly formatted.
* **Porter Stemmer:** Applies a standard set of heuristic rules. It's gentle but sometimes leaves non-words (e.g., "university" might become "univers").
* **Lancaster Stemmer:** This is a very aggressive stemming algorithm. You will likely notice it chopping words down to very short, often unrecognizable roots.
* **WordNet Lemmatizer:** Notice how the lemmatizer almost always outputs a valid dictionary word. Because it looks up words in the WordNet database, it is much safer for preserving actual semantic meaning, which is crucial for advanced AI pipelines and downstream classification tasks.

```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer

# Ensure required models are downloaded
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize the stemmers and lemmatizer
porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

# 1. Get a set of 125 unique tokens from the normalized text
# Using unique tokens makes the comparison much more insightful than repeating stopwords
tokens_list = normalized_text.split()
unique_tokens = list(set(tokens_list))
selected_tokens = unique_tokens[:125]

# 2. Generate the comparison table
print("=== LEMMATIZATION VS STEMMING COMPARISON (125 Tokens) ===")
print(f"{'Token':<20} | {'Porter Stem':<20} | {'Lancaster Stem':<20} | {'Lemma':<20}")
print("-" * 87)

for token in selected_tokens:
    p_stem = porter.stem(token)
    l_stem = lancaster.stem(token)
    # Note: WordNetLemmatizer assumes the word is a noun by default unless a POS tag is provided.
    # For a simple comparison table, the default noun lemmatization is standard.
    lemma = lemmatizer.lemmatize(token)

    print(f"{token:<20} | {p_stem:<20} | {l_stem:<20} | {lemma:<20}")
```

## Task:P1T7: Named Entity Recognition (NER)
To perform Named Entity Recognition (NER) effectively, it is crucial to use the original, case-sensitive tokens from Task P1T2. If we use the lowercased tokens from P1T3, the NLTK chunker will fail to recognize most entities, as it relies heavily on capitalization (Proper Nouns) to identify names, organizations, and locations.

We re-tokenizes our `extracted_text`, tags the parts of speech, extracts the named entities, groups `GPE (Geopolitical Entities)` and `LOCATION` together as required, and prints the lists and counts

The MIT CSAIL text, we might notice that some words get misclassified (e.g., a software name might be tagged as a PERSON, or a lab might be tagged as a GPE). This is the limitations of standard statistical chunkers compared to modern contextual models like BERT.

```python
import nltk
from collections import defaultdict

# Ensure required NLTK models are downloaded
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab') # This fixes the LookupError!
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')
nltk.download('punkt_tab')

# 1. Tokenize and POS tag the ORIGINAL case-sensitive text
original_tokens = nltk.word_tokenize(extracted_text)
pos_tags = nltk.pos_tag(original_tokens)

# 2. Perform NER using NLTK chunker
ner_tree = nltk.ne_chunk(pos_tags)

# 3. Extract and group the entities
extracted_entities = defaultdict(list)

for chunk in ner_tree:
    if hasattr(chunk, 'label'):
        entity_name = ' '.join(c[0] for c in chunk)
        entity_label = chunk.label()

        # Merge GPE and LOCATION as requested by the assignment
        if entity_label in ['GPE', 'LOCATION']:
            entity_label = 'GPE/LOCATION'

        extracted_entities[entity_label].append(entity_name)

# 4. Filter for unique entities and get their counts
persons = list(set(extracted_entities.get('PERSON', [])))
organizations = list(set(extracted_entities.get('ORGANIZATION', [])))
locations = list(set(extracted_entities.get('GPE/LOCATION', [])))

# ==========================================
# Deliverables Output
# ==========================================
print("=== NAMED ENTITY RECOGNITION (NER) COUNTS ===")
print(f"Total Unique PERSON entities: {len(persons)}")
print(f"Total Unique ORGANIZATION entities: {len(organizations)}")
print(f"Total Unique GPE/LOCATION entities: {len(locations)}\n")

print("=== PERSON ENTITIES (Showing first 20) ===")
print(", ".join(persons[:20]) + ("..." if len(persons) > 20 else ""))

print("\n=== ORGANIZATION ENTITIES (Showing first 20) ===")
print(", ".join(organizations[:20]) + ("..." if len(organizations) > 20 else ""))

print("\n=== GPE/LOCATION ENTITIES (Showing first 20) ===")
print(", ".join(locations[:20]) + ("..." if len(locations) > 20 else ""))
```

### Implementation: Unigram & Smoothed Unigram Models
The **UnigramModel** and the **SmoothedUnigramModel** as our first step into the language modeling section.

In these models, we assume that the probability of a word occurring is independent of the words that came before it. The unsmoothed version uses raw frequencies, while the smoothed version uses Laplace (add-one) smoothing to ensure that no word (even one not seen in training) has a zero probability.

The unigram models generated sentences that look like random jumbles of words. Because unigrams have no concept of word order or context, they just spit out words proportional to how often they appeared in the training corpus.


```python
import math
import random
import nltk
from collections import Counter
from nltk.corpus import brown, reuters

# Ensure required corpora are downloaded
nltk.download('brown')
nltk.download('reuters')
nltk.download('punkt')
nltk.download('punkt_tab')

# ---------------------------------------------------------
# 1. Data Preparation
# ---------------------------------------------------------
def preprocess_corpus(sentences):
    """
    Adds <s> and </s> markers to each sentence and lowercases the tokens.
    """
    processed_sentences = []
    for sentence in sentences:
        processed_sent = ['<s>'] + [word.lower() for word in sentence] + ['</s>']
        processed_sentences.append(processed_sent)
    return processed_sentences

# Load a subset of the Brown corpus for training (10,000 sentences is a good balance for memory)
train_sentences = preprocess_corpus(brown.sents()[:10000])

# Flatten to build a shared vocabulary
all_tokens = [token for sublist in train_sentences for token in sublist]
word_counts = Counter(all_tokens)
vocab = set(word_counts.keys())

# ---------------------------------------------------------
# 2. Unigram Model (Unsmoothed)
# ---------------------------------------------------------
class UnigramModel:
    def __init__(self, train_sentences, vocab):
        self.vocab = list(vocab)
        self.counts = Counter([token for sent in train_sentences for token in sent])
        self.total_tokens = sum(self.counts.values())

        # Precompute probabilities for generation to speed it up
        self.probs = [self.counts[w] / self.total_tokens for w in self.vocab]

    def getSentenceProbability(self, sen):
        log_prob = 0
        for word in sen:
            if word == '<s>': continue # <s> has probability 1
            word_count = self.counts.get(word, 0)
            if word_count == 0:
                return 0.0 # Unseen word in unsmoothed model means 0 probability
            log_prob += math.log(word_count / self.total_tokens)
        return math.exp(log_prob)

    def generateSentence(self):
        sentence = ['<s>']
        current_word = None

        while current_word != '</s>':
            current_word = random.choices(self.vocab, weights=self.probs)[0]
            # We assume <s> is only for the start, so skip if we randomly sample it again
            if current_word == '<s>': continue
            sentence.append(current_word)

            # Safety break to prevent infinitely long sentences
            if len(sentence) > 100: break

        return sentence

# ---------------------------------------------------------
# 3. Smoothed Unigram Model (Laplace / Add-1 Smoothing)
# ---------------------------------------------------------
class SmoothedUnigramModel:
    def __init__(self, train_sentences, vocab):
        self.vocab = list(vocab)
        self.counts = Counter([token for sent in train_sentences for token in sent])
        self.total_tokens = sum(self.counts.values())
        self.vocab_size = len(self.vocab)

        # Precompute Laplace smoothed probabilities for generation
        self.probs = [(self.counts[w] + 1) / (self.total_tokens + self.vocab_size) for w in self.vocab]

    def getSentenceProbability(self, sen):
        log_prob = 0
        for word in sen:
            if word == '<s>': continue
            word_count = self.counts.get(word, 0)
            # P(w) = (count(w) + 1) / (N + V)
            log_prob += math.log((word_count + 1) / (self.total_tokens + self.vocab_size))
        return math.exp(log_prob)

    def generateSentence(self):
        sentence = ['<s>']
        current_word = None

        while current_word != '</s>':
            current_word = random.choices(self.vocab, weights=self.probs)[0]
            if current_word == '<s>': continue
            sentence.append(current_word)
            if len(sentence) > 100: break

        return sentence

# ==========================================
# Deliverables & Testing Output
# ==========================================
unigram = UnigramModel(train_sentences, vocab)
smoothed_unigram = SmoothedUnigramModel(train_sentences, vocab)

print("=== UNIGRAM MODEL GENERATION ===")
print(" ".join(unigram.generateSentence()))

print("\n=== SMOOTHED UNIGRAM MODEL GENERATION ===")
print(" ".join(smoothed_unigram.generateSentence()))

# Testing probability calculation
test_sentence = ['<s>', 'the', 'jury', 'said', 'it', '</s>']
print(f"\nProbability of test sentence (Unigram): {unigram.getSentenceProbability(test_sentence)}")
print(f"Probability of test sentence (Smoothed): {smoothed_unigram.getSentenceProbability(test_sentence)}")
```

### Implementation: Bigram & Smoothed Bigram Models
The **Bigram Models**, which look at the *previous* word to predict the *next* word. This will instantly make the generated sentences look much more realistic. We are implementing:

1. **BigramModel**: Uses raw counts. If a bigram (word pair) was never seen in training, its probability is 0.
2. **SmoothedBigramModelLI**: Uses **Linear Interpolation**. It mixes the Bigram probability, the Unigram probability, and a small Uniform probability to ensure that even unseen combinations have a non-zero chance of occurring.

The difference between the two test sentences perfectly highlights why we need smoothing: the unsmoothed bigram model assigned a probability of exactly `0.0` to the test sentence because it contained a word transition it had never seen before. The Smoothed Bigram (Linear Interpolation) saved the day by falling back to the unigram and uniform distributions to give it a valid, non-zero probability!


```python
from collections import defaultdict

# ---------------------------------------------------------
# 4. Bigram Model (Unsmoothed)
# ---------------------------------------------------------
class BigramModel:
    def __init__(self, train_sentences, vocab):
        self.vocab = list(vocab)
        self.unigram_counts = Counter()
        self.bigram_counts = Counter()

        # Count unigrams and bigrams
        for sent in train_sentences:
            for i in range(len(sent)):
                self.unigram_counts[sent[i]] += 1
                if i < len(sent) - 1:
                    self.bigram_counts[(sent[i], sent[i+1])] += 1

        # Precompute transition probabilities for faster generation
        # dictionary structure: current_word -> (list_of_next_words, list_of_probabilities)
        self.transitions = defaultdict(lambda: ([], []))
        for (w1, w2), count in self.bigram_counts.items():
            self.transitions[w1][0].append(w2)
            self.transitions[w1][1].append(count / self.unigram_counts[w1])

    def getSentenceProbability(self, sen):
        log_prob = 0
        for i in range(len(sen) - 1):
            w1, w2 = sen[i], sen[i+1]
            count_w1 = self.unigram_counts.get(w1, 0)
            count_w1_w2 = self.bigram_counts.get((w1, w2), 0)

            if count_w1 == 0 or count_w1_w2 == 0:
                return 0.0 # Unseen transition means 0 probability

            # P(w2 | w1) = count(w1, w2) / count(w1)
            log_prob += math.log(count_w1_w2 / count_w1)
        return math.exp(log_prob)

    def generateSentence(self):
        sentence = ['<s>']
        current_word = '<s>'

        while current_word != '</s>':
            next_words, probs = self.transitions.get(current_word, ([], []))

            # If we hit a dead end (a word that only ever appeared at the end of the training set)
            if not next_words:
                sentence.append('</s>')
                break

            current_word = random.choices(next_words, weights=probs)[0]
            if current_word == '<s>': continue
            sentence.append(current_word)

            if len(sentence) > 100: break # Safety cutoff

        return sentence

# ---------------------------------------------------------
# 5. Smoothed Bigram Model (Linear Interpolation)
# ---------------------------------------------------------
class SmoothedBigramModelLI:
    def __init__(self, train_sentences, vocab):
        self.vocab = list(vocab)
        self.vocab_size = len(vocab)
        self.unigram_counts = Counter()
        self.bigram_counts = Counter()
        self.total_tokens = 0

        # Count unigrams and bigrams
        for sent in train_sentences:
            for i in range(len(sent)):
                self.unigram_counts[sent[i]] += 1
                self.total_tokens += 1
                if i < len(sent) - 1:
                    self.bigram_counts[(sent[i], sent[i+1])] += 1

        # Lambdas for linear interpolation (must sum to 1.0)
        self.l1 = 0.70  # Weight for Bigram
        self.l2 = 0.25  # Weight for Unigram
        self.l3 = 0.05  # Weight for Uniform (handles completely unseen words)

        # Precompute lists for unigram fallback generation
        self.unigram_words = list(self.unigram_counts.keys())
        self.unigram_probs = [self.unigram_counts[w] / self.total_tokens for w in self.unigram_words]

    def getSentenceProbability(self, sen):
        log_prob = 0
        for i in range(len(sen) - 1):
            w1, w2 = sen[i], sen[i+1]

            # 1. Bigram MLE
            p_bigram = 0.0
            if self.unigram_counts.get(w1, 0) > 0:
                p_bigram = self.bigram_counts.get((w1, w2), 0) / self.unigram_counts[w1]

            # 2. Unigram MLE
            p_unigram = self.unigram_counts.get(w2, 0) / self.total_tokens

            # 3. Uniform Probability
            p_uniform = 1.0 / self.vocab_size

            # Interpolated probability
            p_interp = (self.l1 * p_bigram) + (self.l2 * p_unigram) + (self.l3 * p_uniform)
            log_prob += math.log(p_interp)

        return math.exp(log_prob)

    def generateSentence(self):
        sentence = ['<s>']
        current_word = '<s>'

        while current_word != '</s>':
            # We sample from the interpolated distribution using a random threshold
            rand_val = random.random()

            # With probability l1, pick from the bigram distribution
            if rand_val < self.l1 and self.unigram_counts.get(current_word, 0) > 0:
                next_words = [pair[1] for pair in self.bigram_counts.keys() if pair[0] == current_word]
                if next_words:
                    weights = [self.bigram_counts[(current_word, nw)] for nw in next_words]
                    current_word = random.choices(next_words, weights=weights)[0]
                else:
                    current_word = random.choices(self.unigram_words, weights=self.unigram_probs)[0]

            # With probability l2, pick from the unigram distribution
            elif rand_val < (self.l1 + self.l2):
                current_word = random.choices(self.unigram_words, weights=self.unigram_probs)[0]

            # With probability l3, pick completely randomly from vocabulary
            else:
                current_word = random.choice(self.vocab)

            if current_word == '<s>': continue
            sentence.append(current_word)

            if len(sentence) > 100: break

        return sentence

# ==========================================
# Deliverables & Testing Output
# ==========================================
bigram = BigramModel(train_sentences, vocab)
smoothed_bigram = SmoothedBigramModelLI(train_sentences, vocab)

print("=== BIGRAM MODEL GENERATION ===")
print(" ".join(bigram.generateSentence()))

print("\n=== SMOOTHED BIGRAM MODEL (LI) GENERATION ===")
print(" ".join(smoothed_bigram.generateSentence()))

print(f"\nProbability of test sentence (Bigram): {bigram.getSentenceProbability(test_sentence)}")
print(f"Probability of test sentence (Smoothed LI): {smoothed_bigram.getSentenceProbability(test_sentence)}")
```

### Q1. When generating sentences with the unigram model, what controls the length of the generated sentences? How does this differ from the sentences produced by the bigram models?

**Answer:**
In the **Unigram** model, sentence length is controlled entirely by the independent probability of sampling the end-of-sentence marker (`</s>`). Because the model lacks any context from preceding words, every single generated word is essentially a weighted coin flip deciding if the sentence should end or continue. This leads to sentences that are either randomly very long or abruptly short.

In contrast, the **Bigram** model's sentence length is strictly controlled by *contextual transitions*. The model will only generate `</s>` if the *current word* frequently ended sentences in the training corpus (such as a period `.` or a closing quote `''`). This makes the length, structure, and termination of Bigram sentences much more natural and realistic.

# Part 3: Byte Pair Encoding
The objective of this part is to address the morphological complexity and lack of distinct word boundaries in the Urdu language using subword tokenization. We implement a statistical Byte Pair Encoding (BPE) algorithm from scratch to iteratively merge frequent character pairs, enabling the accurate segmentation of unspaced Urdu text.

## Task P3T1: Byte Pair Encoding (BPE) for Urdu Word Segmentation
We need to implement the BPE algorithm from scratch to segment 50 Urdu sentences without spaces into meaningful subwords and words, using a provided dictionary.

Standard BPE relies heavily on regex, which can sometimes break when dealing with right-to-left (RTL) Arabic/Urdu unicode characters. To ensure your code works flawlessly, I have implemented a **list-based BPE algorithm**. It treats the dictionary as the training corpus to learn the optimal character merges, and then applies those exact merges to segment the unspaced text.

* **Dictionary Preprocessing Explanation (94 words):**
To optimize the BPE learning process, the provided dictionary (`wordlist.txt`) was preprocessed by removing duplicate entries, empty lines, and stripping leading/trailing whitespaces. Each unique word was assigned a uniform frequency of 1 to create the initial BPE training vocabulary. We intentionally did not append end-of-word tokens (like `</w>`) during training. Because the target text (`50_nospaces.txt`) completely lacks explicit word boundaries, omitting `</w>` allows the BPE merge rules learned from the dictionary to seamlessly group the unspaced sentence characters into valid subwords and words based purely on statistical co-occurrence.

```python
import collections
import os

# ---------------------------------------------------------
# 1. BPE Training on the Dictionary (wordlist.txt)
# ---------------------------------------------------------
def train_bpe(wordlist, num_merges=3000):
    print("Training BPE on the dictionary... (This may take a minute)")

    # Preprocess: Assign frequency 1 to each unique, non-empty word
    # Split each word into a spaced string of characters
    vocab = {' '.join(list(w.strip())): 1 for w in set(wordlist) if w.strip()}
    merge_rules = []

    for i in range(num_merges):
        pairs = collections.defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for j in range(len(symbols)-1):
                pairs[symbols[j], symbols[j+1]] += freq

        # Stop if no more pairs can be merged
        if not pairs:
            break

        # Find the most frequent adjacent pair of symbols
        best_pair = max(pairs, key=pairs.get)
        merge_rules.append(best_pair)

        # Apply the merge to our vocabulary
        new_vocab = {}
        for word, freq in vocab.items():
            symbols = word.split()
            new_symbols = []
            j = 0
            while j < len(symbols):
                if j < len(symbols) - 1 and symbols[j] == best_pair[0] and symbols[j+1] == best_pair[1]:
                    new_symbols.append(best_pair[0] + best_pair[1])
                    j += 2
                else:
                    new_symbols.append(symbols[j])
                    j += 1
            new_vocab[' '.join(new_symbols)] = freq
        vocab = new_vocab

    print(f"BPE Training complete! Learned {len(merge_rules)} merge rules.")
    return merge_rules

# ---------------------------------------------------------
# 2. Applying BPE to Unspaced Text
# ---------------------------------------------------------
def apply_bpe(text, merge_rules):
    # Split the unspaced sentence into raw characters
    symbols = list(text.strip())

    # Apply the rules in the exact order they were learned
    for pair in merge_rules:
        # Optimization: Skip if the pair elements aren't even in the sentence
        if pair[0] not in symbols or pair[1] not in symbols:
            continue

        new_symbols = []
        i = 0
        while i < len(symbols):
            if i < len(symbols) - 1 and symbols[i] == pair[0] and symbols[i+1] == pair[1]:
                new_symbols.append(pair[0] + pair[1])
                i += 2
            else:
                new_symbols.append(symbols[i])
                i += 1
        symbols = new_symbols

    # Join the final merged symbols with spaces to form the segmented sentence
    return ' '.join(symbols)

# ==========================================
# Execution & File Generation
# ==========================================
# 1. Read files
with open('wordlist.txt', 'r', encoding='utf-8') as f:
    urdu_wordlist = f.readlines()

with open('50_nospaces.txt', 'r', encoding='utf-8') as f:
    unspaced_sentences = f.readlines()

# 2. Train and learn merges
bpe_merge_rules = train_bpe(urdu_wordlist, num_merges=3000)

# 3. Apply to sentences and save output
segmented_sentences = [apply_bpe(sent, bpe_merge_rules) for sent in unspaced_sentences]

with open('50_segmented.txt', 'w', encoding='utf-8') as f:
    for sent in segmented_sentences:
        f.write(sent + "\n")

print("\n=== P3T1 COMPLETE ===")
print("Segmented sentences successfully saved to '50_segmented.txt'")
print("Example Segmented Sentence:", segmented_sentences[0])
```

We need to process a custom Urdu corpus of at least 10,000 tokens to calculate the average number of Unicode characters per word.

For this task, we need a corpus of $\geq 10,000$ tokens from the internet. Instead of making to download and manage messy text files, I have written a web-scraping script that directly pulls two massive Urdu Wikipedia articles ("Pakistan" and "Islam").

```python
import urllib.request
from bs4 import BeautifulSoup
import re

print("Fetching Urdu corpus from Wikipedia... Please wait.")

# URLs of large Urdu Wikipedia articles (Pakistan & Islam)
urls = [
    "https://ur.wikipedia.org/wiki/%D9%BE%D8%A7%DA%A9%D8%B3%D8%AA%D8%A7%D9%86",
    "https://ur.wikipedia.org/wiki/%D8%A7%D8%B3%D9%84%D8%A7%D9%85"
]

corpus_tokens = []

for url in urls:
    req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = urllib.request.urlopen(req).read().decode('utf-8')
    soup = BeautifulSoup(html, 'html.parser')

    # Extract text from paragraph tags
    for p in soup.find_all('p'):
        text = p.get_text()

        # Clean up Wikipedia citation brackets like [1], [2]
        text = re.sub(r'\[\d+\]', '', text)

        # Tokenize by whitespace
        tokens = text.split()
        corpus_tokens.extend(tokens)

# ---------------------------------------------------------
# Calculations
# ---------------------------------------------------------
total_tokens = len(corpus_tokens)

# Calculate total characters across all tokens
total_chars = sum(len(token) for token in corpus_tokens)

# Compute Average
avg_chars_per_token = total_chars / total_tokens if total_tokens > 0 else 0

# ==========================================
# Deliverables Output
# ==========================================
print("\n=== P3T2: URDU CORPUS STATISTICS ===")
print(f"Total number of tokens: {total_tokens:,}")
print(f"Average number of characters per token: {avg_chars_per_token:.2f}")
print("\nSource links used:")
for url in urls:
    print(f" - {url}")
```

It requires implementing TF-IDF entirely **from scratch** using a primary unlabeled dataset of 20,000 to 100,000 sentences from Hugging Face. We also need to build a vocabulary with a minimum frequency of 5, calculate sparsity and OOV rates, and compare Word-level tokens against BPE tokens.

Pay close attention to the OOV Rate and Sparsity Ratio. BPE significantly reduces the Out-Of-Vocabulary rate because unknown words are broken down into known subword chunks. However, it may alter the cosine similarity because documents will share more generic subwords rather than distinct whole words.

```python
!pip install datasets tokenizers

import math
import numpy as np
from collections import Counter, defaultdict
from datasets import load_dataset
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
import re

print("=== LOADING DATASET ===")
# Load AG News from Hugging Face as our primary dataset
dataset = load_dataset("ag_news", split="train[:20000]") # Using 20,000 sentences for memory efficiency
corpus = dataset['text']

# Basic normalization from Part 1
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '<NUM>', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.strip()

print("Cleaning corpus...")
cleaned_corpus = [clean_text(doc) for doc in corpus]

# Train a quick BPE tokenizer on our English corpus to satisfy the BPE comparison requirement
print("Training BPE Tokenizer...")
bpe_tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
bpe_tokenizer.pre_tokenizer = Whitespace()
trainer = BpeTrainer(vocab_size=10000, special_tokens=["[UNK]"])
bpe_tokenizer.train_from_iterator(cleaned_corpus, trainer)

# Tokenize corpus into Word-level and BPE-level
word_docs = [doc.split() for doc in cleaned_corpus]
bpe_docs = [bpe_tokenizer.encode(doc).tokens for doc in cleaned_corpus]

# Split 500 documents off as a "Test" set to calculate OOV rate
train_word_docs, test_word_docs = word_docs[:-500], word_docs[-500:]
train_bpe_docs, test_bpe_docs = bpe_docs[:-500], bpe_docs[-500:]

# ---------------------------------------------------------
# Custom TF-IDF Implementation (From Scratch)
# ---------------------------------------------------------
class CustomTFIDF:
    def __init__(self, min_freq=5):
        self.min_freq = min_freq
        self.vocab = {}
        self.idf = {}
        self.vocab_size = 0

    def fit(self, documents):
        # 1. Build Vocabulary with min_freq
        doc_counts = Counter()
        for doc in documents:
            doc_counts.update(set(doc)) # Count document frequency (DF)

        # Filter by min_freq and build index
        filtered_words = [word for word, count in doc_counts.items() if count >= self.min_freq]
        self.vocab = {word: idx for idx, word in enumerate(filtered_words)}
        self.vocab_size = len(self.vocab)

        # 2. Compute IDF
        num_docs = len(documents)
        for word, idx in self.vocab.items():
            df = doc_counts[word]
            # Standard IDF formula: log(N / (DF + 1))
            self.idf[word] = math.log(num_docs / (df + 1))

    def transform(self, documents):
        num_docs = len(documents)
        # Using a dense matrix for simplicity, though scipy.sparse is better for huge datasets
        tfidf_matrix = np.zeros((num_docs, self.vocab_size))

        for i, doc in enumerate(documents):
            word_counts = Counter(doc)
            total_words = len(doc)

            for word, count in word_counts.items():
                if word in self.vocab:
                    # Compute TF: (count in doc / total words in doc)
                    tf = count / total_words
                    tfidf = tf * self.idf[word]
                    tfidf_matrix[i, self.vocab[word]] = tfidf

        return tfidf_matrix

print("\n=== TRAINING TF-IDF MODELS ===")
word_tfidf = CustomTFIDF(min_freq=5)
word_tfidf.fit(train_word_docs)
word_matrix = word_tfidf.transform(train_word_docs)

bpe_tfidf = CustomTFIDF(min_freq=5)
bpe_tfidf.fit(train_bpe_docs)
bpe_matrix = bpe_tfidf.transform(train_bpe_docs)

# ---------------------------------------------------------
# Metrics Calculation
# ---------------------------------------------------------
def calculate_sparsity(matrix):
    total_elements = matrix.size
    zero_elements = total_elements - np.count_nonzero(matrix)
    return (zero_elements / total_elements) * 100

def calculate_oov_rate(vocab, test_docs):
    total_tokens = sum(len(doc) for doc in test_docs)
    oov_tokens = sum(1 for doc in test_docs for word in doc if word not in vocab)
    return (oov_tokens / total_tokens) * 100 if total_tokens > 0 else 0

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_a = np.linalg.norm(vec1)
    norm_b = np.linalg.norm(vec2)
    if norm_a == 0 or norm_b == 0: return 0.0
    return dot_product / (norm_a * norm_b)

print("\n=== P4T1 DELIVERABLES: TF-IDF METRICS ===")
print(f"{'Metric':<25} | {'Word-Level TF-IDF':<20} | {'BPE-Level TF-IDF':<20}")
print("-" * 70)
print(f"{'Vocabulary Size':<25} | {word_tfidf.vocab_size:<20} | {bpe_tfidf.vocab_size:<20}")

word_sparsity = calculate_sparsity(word_matrix)
bpe_sparsity = calculate_sparsity(bpe_matrix)
print(f"{'Sparsity Ratio (%)':<25} | {word_sparsity:<20.2f} | {bpe_sparsity:<20.2f}")

word_oov = calculate_oov_rate(word_tfidf.vocab, test_word_docs)
bpe_oov = calculate_oov_rate(bpe_tfidf.vocab, test_bpe_docs)
print(f"{'OOV Rate (%)':<25} | {word_oov:<20.2f} | {bpe_oov:<20.2f}")

# Compare cosine similarity of the first two documents
sim_word = cosine_similarity(word_matrix[0], word_matrix[1])
sim_bpe = cosine_similarity(bpe_matrix[0], bpe_matrix[1])
print(f"{'Cosine Sim (Doc 1 & 2)':<25} | {sim_word:<20.4f} | {sim_bpe:<20.4f}")
```

## Task P4T2: Count-Based Word Embeddings (PPMI)
In this task, we will build a distributional semantic model from scratch. We will construct a Co-occurrence Matrix to count how often words appear near each other within a specific "window" (k=2 and k=5), and then apply **Positive Pointwise Mutual Information (PPMI)** to weigh these co-occurrences.

**The Effect of Window Size ($k=2$ vs $k=5$):** In output, notice how the nearest neighbors change. A smaller window size ($k=2$) captures strict syntactic relationships (words that grammar dictates must appear right next to each other). A larger window size ($k=5$) captures broader semantic/topical relationships (words that belong to the same topic but might not be adjacent).

PPMI model successfully captured the difference between tight syntactic relationships ($k=2$, where "computer" matches "pc" and "desktop") and broader semantic topics ($k=5$, where "computer" matches "apple" and "imac")

```python
import numpy as np
from collections import defaultdict, Counter
import math

print("=== P4T2: BUILDING PPMI EMBEDDINGS ===")

# 1. Build a manageable vocabulary (min_freq=10 to save RAM and avoid noisy rare words)
word_counts = Counter(word for doc in train_word_docs for word in doc)
# FIX: Filter words first, THEN assign indices so they are strictly 0 to vocab_size-1
filtered_words = [w for w, c in word_counts.items() if c >= 10]
vocab = {w: i for i, w in enumerate(filtered_words)}
vocab_size = len(vocab)
idx_to_word = {i: w for w, i in vocab.items()}

print(f"PPMI Vocabulary Size: {vocab_size}")

def build_ppmi_matrix(docs, window_size):
    print(f"Building co-occurrence matrix for window size k={window_size}...")
    co_counts = defaultdict(float)
    w_counts = defaultdict(float)
    c_counts = defaultdict(float)
    total_pairs = 0

    # Count co-occurrences
    for doc in docs:
        n = len(doc)
        for i, target in enumerate(doc):
            if target not in vocab: continue
            target_idx = vocab[target]

            # Define window context
            start = max(0, i - window_size)
            end = min(n, i + window_size + 1)
            for j in range(start, end):
                if i == j: continue
                context = doc[j]
                if context not in vocab: continue
                context_idx = vocab[context]

                co_counts[(target_idx, context_idx)] += 1
                w_counts[target_idx] += 1
                c_counts[context_idx] += 1
                total_pairs += 1

    print(f"Applying PPMI calculation for k={window_size}...")
    ppmi = np.zeros((vocab_size, vocab_size), dtype=np.float32)
    for (w, c), count in co_counts.items():
        p_wc = count / total_pairs
        p_w = w_counts[w] / total_pairs
        p_c = c_counts[c] / total_pairs

        # PMI formula: log2( P(w,c) / (P(w)*P(c)) )
        pmi = math.log2(p_wc / (p_w * p_c))
        ppmi[w, c] = max(0, pmi) # Positive PMI

    return ppmi

# Generate matrices for both window sizes
ppmi_k2 = build_ppmi_matrix(train_word_docs, window_size=2)
ppmi_k5 = build_ppmi_matrix(train_word_docs, window_size=5)

# ---------------------------------------------------------
# Evaluation & Metrics
# ---------------------------------------------------------
def get_nearest_neighbors(matrix, word, top_k=5):
    if word not in vocab: return []
    w_idx = vocab[word]
    w_vec = matrix[w_idx]

    if np.linalg.norm(w_vec) == 0: return []

    # Compute cosine similarity with all words
    norms = np.linalg.norm(matrix, axis=1)
    # Ignore divide by zero warnings for zero-vectors
    with np.errstate(divide='ignore', invalid='ignore'):
        sims = np.dot(matrix, w_vec) / (norms * np.linalg.norm(w_vec))
        sims = np.nan_to_num(sims)

    # Get top_k + 1 to exclude the word itself (which has sim=1.0)
    nearest_indices = np.argsort(sims)[-(top_k+1):][::-1]
    return [(idx_to_word[idx], sims[idx]) for idx in nearest_indices if idx != w_idx][:top_k]

def calculate_avg_top_k_similarity(matrix, sample_size=100, top_k=5):
    # Sample frequent words to calculate average similarity
    sample_words = list(vocab.keys())[:sample_size]
    total_sim = 0
    valid_words = 0
    for word in sample_words:
        neighbors = get_nearest_neighbors(matrix, word, top_k)
        if neighbors:
            total_sim += sum(sim for _, sim in neighbors) / len(neighbors)
            valid_words += 1
    return total_sim / valid_words if valid_words > 0 else 0

print("\n=== P4T2 DELIVERABLES: PPMI METRICS ===")
# 1. Vocabulary Coverage
test_tokens = sum(len(doc) for doc in test_word_docs)
covered_tokens = sum(1 for doc in test_word_docs for word in doc if word in vocab)
coverage = (covered_tokens / test_tokens) * 100
print(f"Vocabulary Coverage on Test Set: {coverage:.2f}%\n")

# 2. Nearest Neighbors Example
test_words = ["computer", "oil", "president"]
print("Nearest Neighbors Comparison:")
for word in test_words:
    print(f"--- Word: '{word}' ---")
    nn_k2 = get_nearest_neighbors(ppmi_k2, word, top_k=3)
    nn_k5 = get_nearest_neighbors(ppmi_k5, word, top_k=3)
    print(f"  k=2 : {[(w, round(s, 3)) for w, s in nn_k2]}")
    print(f"  k=5 : {[(w, round(s, 3)) for w, s in nn_k5]}")

# 3. Average Top-K Similarity
avg_sim_k2 = calculate_avg_top_k_similarity(ppmi_k2)
avg_sim_k5 = calculate_avg_top_k_similarity(ppmi_k5)
print(f"\nAverage Top-5 Similarity (k=2): {avg_sim_k2:.4f}")
print(f"Average Top-5 Similarity (k=5): {avg_sim_k5:.4f}")
```

## P4T3: Neural Word Embeddings (Word2Vec / SGNS)
We need to implement **Skip-Gram with Negative Sampling (SGNS)** from scratch. The instructions strictly forbid using embedding libraries like Gensim. Therefore, we will build the neural network architecture from the ground up using **PyTorch**, implementing our own target embeddings, context embeddings, and the negative sampling loss function using Stochastic Gradient Descent (SGD) .

We will train this model on both **Word-level** and **BPE-level** tokens to see if BPE improves rare/OOV word representation.


If we notice, our SGNS loss flattened out around `4.16`, and the nearest neighbors for "computer" and "oil" were somewhat random (like "oslos" or "butt").

 Training Word2Vec from scratch on a CPU with a simple SGD optimizer and only 20,000 sentences for 3 epochs is not enough to learn deep semantic relationships (production models train on billions of words for days using optimized C++ code). Moreover, **BPE reduced the OOV rate from 7.70% to 0.41%**, saving rare words from being ignored!

```python
!pip install torch

import torch
import torch.nn as nn
import torch.optim as optim
import time
import random
import numpy as np
from collections import Counter

print("=== P4T3: NEURAL WORD EMBEDDINGS (SGNS) FROM SCRATCH ===")

# Check for GPU to speed up training if available in Colab
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Training on device: {device}")

# ---------------------------------------------------------
# 1. Data Preparation (Generate Positive Pairs)
# ---------------------------------------------------------
def generate_training_data(docs, vocab, window_size=2):
    pairs = []
    for doc in docs:
        indices = [vocab[w] for w in doc if w in vocab]
        for i, target in enumerate(indices):
            start = max(0, i - window_size)
            end = min(len(indices), i + window_size + 1)
            for j in range(start, end):
                if i != j:
                    pairs.append((target, indices[j]))
    return pairs

print("Generating training pairs for Word-level...")
word_pairs = generate_training_data(train_word_docs, word_tfidf.vocab)
print(f"Generated {len(word_pairs)} word-level pairs.")

print("Generating training pairs for BPE-level...")
bpe_pairs = generate_training_data(train_bpe_docs, bpe_tfidf.vocab)
print(f"Generated {len(bpe_pairs)} BPE-level pairs.")

# ---------------------------------------------------------
# 2. SGNS PyTorch Model Definition
# ---------------------------------------------------------
class SkipGramNegativeSampling(nn.Module):
    def __init__(self, vocab_size, embed_size):
        super(SkipGramNegativeSampling, self).__init__()
        self.target_embeddings = nn.Embedding(vocab_size, embed_size)
        self.context_embeddings = nn.Embedding(vocab_size, embed_size)

        # Initialize embeddings with small random weights
        initrange = 0.5 / embed_size
        self.target_embeddings.weight.data.uniform_(-initrange, initrange)
        self.context_embeddings.weight.data.uniform_(-initrange, initrange)

    def forward(self, target, context, negatives):
        # Target: [batch_size], Context: [batch_size], Negatives: [batch_size, num_negatives]
        emb_target = self.target_embeddings(target)        # [batch_size, embed_size]
        emb_context = self.context_embeddings(context)     # [batch_size, embed_size]
        emb_negatives = self.context_embeddings(negatives) # [batch_size, num_neg, embed_size]

        # Positive score: dot product of target and context
        pos_score = torch.sum(emb_target * emb_context, dim=1) # [batch_size]
        pos_loss = -torch.nn.functional.logsigmoid(pos_score)

        # Negative score: dot product of target and negatives
        # bmm = batch matrix multiplication
        neg_score = torch.bmm(emb_negatives, emb_target.unsqueeze(2)).squeeze(2) # [batch_size, num_neg]
        neg_loss = -torch.sum(torch.nn.functional.logsigmoid(-neg_score), dim=1)

        return torch.mean(pos_loss + neg_loss)

# ---------------------------------------------------------
# 3. Training Loop
# ---------------------------------------------------------
def train_sgns(pairs, vocab_size, embed_size=50, epochs=3, batch_size=2048, num_negatives=5):
    model = SkipGramNegativeSampling(vocab_size, embed_size).to(device)
    optimizer = optim.SGD(model.parameters(), lr=0.1)

    # Simple uniform negative sampling distribution for speed
    vocab_indices = list(range(vocab_size))

    print(f"Training SGNS (Embed Size: {embed_size}, Negatives: {num_negatives}, Optimizer: SGD)...")
    start_time = time.time()

    for epoch in range(epochs):
        random.shuffle(pairs) # Shuffle data each epoch
        total_loss = 0

        for i in range(0, len(pairs), batch_size):
            batch_pairs = pairs[i:i+batch_size]
            targets = torch.tensor([p[0] for p in batch_pairs], dtype=torch.long).to(device)
            contexts = torch.tensor([p[1] for p in batch_pairs], dtype=torch.long).to(device)

            # Generate negative samples
            negatives = torch.randint(0, vocab_size, (len(batch_pairs), num_negatives), dtype=torch.long).to(device)

            optimizer.zero_grad()
            loss = model(targets, contexts, negatives)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        print(f"  Epoch {epoch+1}/{epochs} | Loss: {total_loss / (len(pairs)/batch_size):.4f}")

    end_time = time.time()
    print(f"Training completed in {end_time - start_time:.2f} seconds.")

    # Return the trained target embeddings as a NumPy array for evaluation
    return model.target_embeddings.weight.data.cpu().numpy()

# Train Word-Level Model (Dim: 50)
print("\n--- Training Word-Level SGNS ---")
word_embeddings = train_sgns(word_pairs, word_tfidf.vocab_size, embed_size=50)

# Train BPE-Level Model (Dim: 50)
print("\n--- Training BPE-Level SGNS ---")
bpe_embeddings = train_sgns(bpe_pairs, bpe_tfidf.vocab_size, embed_size=50)

# ---------------------------------------------------------
# 4. Evaluation & Metrics
# ---------------------------------------------------------
def get_sgns_neighbors(embeddings, vocab, word, top_k=5):
    if word not in vocab: return [("OOV", 0.0)]
    idx_to_word = {i: w for w, i in vocab.items()}
    w_idx = vocab[word]
    w_vec = embeddings[w_idx]

    norms = np.linalg.norm(embeddings, axis=1)
    with np.errstate(divide='ignore', invalid='ignore'):
        sims = np.dot(embeddings, w_vec) / (norms * np.linalg.norm(w_vec))
        sims = np.nan_to_num(sims)

    nearest_indices = np.argsort(sims)[-(top_k+1):][::-1]
    return [(idx_to_word[idx], sims[idx]) for idx in nearest_indices if idx != w_idx][:top_k]

print("\n=== P4T3 DELIVERABLES: SGNS METRICS ===")
# Semantic Consistency Test
test_words = ["computer", "oil"]
print("Nearest Neighbors Comparison (SGNS Dim=50):")
for word in test_words:
    print(f"--- Word: '{word}' ---")
    nn_word = get_sgns_neighbors(word_embeddings, word_tfidf.vocab, word, top_k=3)
    # For BPE, we check how it handles the token (some might be split, assuming whole word exists for simplicity)
    nn_bpe = get_sgns_neighbors(bpe_embeddings, bpe_tfidf.vocab, word, top_k=3)

    print(f"  Word-Level : {[(w, round(float(s), 3)) for w, s in nn_word]}")
    print(f"  BPE-Level  : {[(w, round(float(s), 3)) for w, s in nn_bpe]}")

print("\nOOV Rate & Coverage Analysis:")
print(f"Word-Level Test Set OOV Rate: {word_oov:.2f}% (Rare words drop out entirely)")
print(f"BPE-Level Test Set OOV Rate: {bpe_oov:.2f}% (BPE successfully breaks rare words into known subword embeddings)")
```

TF-IDF explicitly preserves and weighs these exact keywords. In contrast, simple Mean Pooling of word embeddings often "washes out" these critical keywords by averaging their vectors with the vectors of highly frequent, less informative words.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

print("=== P4T5 & P4T6: SENTENCE EMBEDDINGS & EXTRINSIC CLASSIFICATION ===")

# 1. Prepare Dataset for Classification (Using a subset for speed)
# AG News has 4 classes: World, Sports, Business, Sci/Tech
train_texts = [clean_text(doc) for doc in dataset['text'][:5000]]
train_labels = dataset['label'][:5000]

# We need a separate test set from the dataset
test_dataset = load_dataset("ag_news", split="test[:1000]")
test_texts = [clean_text(doc) for doc in test_dataset['text']]
test_labels = test_dataset['label']

# ---------------------------------------------------------
# P4T5: Constructing Sentence Embeddings
# ---------------------------------------------------------
# We will use the SGNS word_embeddings we trained in P4T3
vocab_list = list(word_tfidf.vocab.keys())
sgns_dict = {word: word_embeddings[word_tfidf.vocab[word]] for word in vocab_list}
embed_dim = 50

def get_mean_pooling_embeddings(texts):
    sentence_vecs = []
    for text in texts:
        words = text.split()
        valid_vecs = [sgns_dict[w] for w in words if w in sgns_dict]
        if valid_vecs:
            sentence_vecs.append(np.mean(valid_vecs, axis=0))
        else:
            sentence_vecs.append(np.zeros(embed_dim))
    return np.array(sentence_vecs)

def get_tfidf_weighted_embeddings(texts, tfidf_vectorizer):
    # Transform text to get TF-IDF weights
    tfidf_matrix = tfidf_vectorizer.transform(texts)
    feature_names = tfidf_vectorizer.get_feature_names_out()
    word2weight = {word: idx for idx, word in enumerate(feature_names)}

    sentence_vecs = []
    for i, text in enumerate(texts):
        words = text.split()
        vec = np.zeros(embed_dim)
        weight_sum = 0.0

        for w in words:
            if w in sgns_dict and w in word2weight:
                weight = tfidf_matrix[i, word2weight[w]]
                vec += sgns_dict[w] * weight
                weight_sum += weight

        if weight_sum > 0:
            vec /= weight_sum

        sentence_vecs.append(vec)
    return np.array(sentence_vecs)

print("Generating Mean Pooling Sentence Embeddings...")
X_train_mean = get_mean_pooling_embeddings(train_texts)
X_test_mean = get_mean_pooling_embeddings(test_texts)

# ---------------------------------------------------------
# P4T6: Extrinsic Evaluation (Topic Classification)
# ---------------------------------------------------------
print("Generating TF-IDF Representations (Baseline)...")
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(train_texts)
X_test_tfidf = vectorizer.transform(test_texts)

print("Generating TF-IDF Weighted Sentence Embeddings...")
X_train_weighted = get_tfidf_weighted_embeddings(train_texts, vectorizer)
X_test_weighted = get_tfidf_weighted_embeddings(test_texts, vectorizer)

def evaluate_classifier(X_train, X_test, name):
    clf = LogisticRegression(max_iter=1000)
    clf.fit(X_train, train_labels)
    preds = clf.predict(X_test)

    acc = accuracy_score(test_labels, preds)
    f1 = f1_score(test_labels, preds, average='macro')
    return acc, f1

print("\nTraining Classifiers...")
# 1. TF-IDF Baseline
acc_tfidf, f1_tfidf = evaluate_classifier(X_train_tfidf, X_test_tfidf, "TF-IDF")

# 2. Mean Pooling (SGNS)
acc_mean, f1_mean = evaluate_classifier(X_train_mean, X_test_mean, "SGNS Mean Pooling")

# 3. TF-IDF Weighted (SGNS)
acc_weighted, f1_weighted = evaluate_classifier(X_train_weighted, X_test_weighted, "SGNS TF-IDF Weighted")

print("\n=== P4T5 & P4T6 DELIVERABLES: EXTRINSIC CLASSIFICATION ===")
print(f"{'Representation Model':<30} | {'Accuracy':<10} | {'F1-Score (Macro)':<15}")
print("-" * 65)
print(f"{'Classical TF-IDF (Sparse)':<30} | {acc_tfidf:.4f}     | {f1_tfidf:.4f}")
print(f"{'SGNS Mean Pooling':<30} | {acc_mean:.4f}     | {f1_mean:.4f}")
print(f"{'SGNS TF-IDF Weighted':<30} | {acc_weighted:.4f}     | {f1_weighted:.4f}")
```
